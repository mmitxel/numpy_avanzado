```python
import numpy as np
```

```python
# ejercicio 1 - Dado el arreglo `a = np.array([1, 2, 3, 4])`, determina si ninguno de los elementos es cero.

a = np.array([1, 2, 3, 4])

def find_zeros(arr):
    return np.where(arr == 0)[0].size > 0

print(f'{"Sí" if find_zeros(a) else "No"} hay algún cero en el array "a"')
```

```python
# ejercicio 2 - Dado el arreglo `a = np.array([0, 0, 2, 0])`, determina si contiene algún elemento que no sea cero.

a = np.array([0, 0, 2, 0])

def find_nonzeros(arr):
    return np.nonzero(arr)[0].size > 0

print(f'{"Sí" if find_nonzeros(a) else "No"} hay algún elemento que no sea cero en el array "a"')
```

```python
# 3. Analiza el arreglo `a = np.array([1, 2, np.nan, np.inf])` y determina si contiene valores que sean infinitos
# o valores no numéricos (NaN).

a = np.array([1, 2, np.nan, np.inf])

def find_nan(arr):
    return np.any(np.isnan(arr))

def find_inf(arr):
    return np.any(np.isinf(arr))

print(f'{"Sí" if find_nan(a) or find_inf(a) else "No"} hay algún elemento NaN o infinito en el array "a"')
```

```python
# 4. Dado el arreglo `a = np.array([1, 2, np.inf])`, determina si contiene valores infinitos.
a = np.array([1, 2, np.inf])

def find_inf(arr):
    return np.any(np.isinf(arr))
print(f'{"Sí" if find_inf(a) else "No"} hay algún elemento infinito en el array "a"')
```

```python
# 5. Dado el arreglo `a = np.array([1, np.nan, 3])`, determina si contiene valores NaN. 
a = np.array([1, np.nan, 3])

def find_nan(arr):
    return np.any(np.isnan(arr))
print(f'{"Sí" if find_nan(a) else "No"} hay algún elemento NaN en el array "a"')
```

```python
# 6. Dados los arreglos `a = np.array([1, 3, 5])` y `b = np.array([2, 3, 4])`, realiza una comparación elemento
# a elemento contemplando mayor, mayor o igual, menor y menor o igual.

a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

condiciones = [a > b, a < b, a == b]
resultados = [" mayor que ", " menor que ", " igual a "]
comparaciones = np.select(condiciones, resultados, default="")

print(a.astype(str)+comparaciones+b.astype(str))
```
